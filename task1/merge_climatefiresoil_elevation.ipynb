````python
import pandas as pd
import numpy as np

# Load your data
df = pd.read_csv(r"D:\S3\rapport data_mining\newdata\final_complete_dataset.csv")

# ============================================================================
# METHOD 1: CREATE SEASONAL AVERAGES (RECOMMENDED)
# ============================================================================

def create_seasonal_features(df):
    """
    Group monthly climate data into seasonal averages.
    
    Seasons:
    - Winter: Dec, Jan, Feb
    - Spring: Mar, Apr, May
    - Summer: Jun, Jul, Aug
    - Autumn: Sep, Oct, Nov
    """
    
    # Define month groupings for each season
    seasons = {
        'winter': ['jan', 'feb', 'dec'],
        'spring': ['mar', 'apr', 'may'],
        'summer': ['jun', 'jul', 'aug'],
        'autumn': ['sep', 'oct', 'nov']
    }
    
    # For each climate variable (tmin, tmax, prec)
    for var_type in ['tmin', 'tmax', 'prec']:
        
        for season_name, months in seasons.items():
            # Get columns for this variable and season
            season_cols = [f"{var_type}_{month}" for month in months 
                          if f"{var_type}_{month}" in df.columns]
            
            if season_cols:
                # Calculate seasonal average
                df[f"{var_type}_{season_name}"] = df[season_cols].mean(axis=1)
                
                print(f"✅ Created {var_type}_{season_name} from {len(season_cols)} months")
    
    return df


# Apply seasonal grouping
df_seasonal = create_seasonal_features(df.copy())

# Save with seasonal features
df_seasonal.to_csv(r"D:\S3\rapport data_mining\newdata\climate_fire_soil_seasonal.csv", index=False)
print("\n✅ Seasonal features added!")

# ============================================================================
# METHOD 2: DROP MONTHLY AND KEEP ONLY SEASONAL (CLEANER)
# ============================================================================

def replace_monthly_with_seasonal(df):
    """
    Replace monthly data with seasonal averages (reduces features).
    """
    
    # Create seasonal features
    df = create_seasonal_features(df)
    
    # Drop all monthly columns
    monthly_cols = [col for col in df.columns 
                   if any(month in col for month in 
                         ['jan', 'feb', 'mar', 'apr', 'may', 'jun',
                          'jul', 'aug', 'sep', 'oct', 'nov', 'dec'])]
    
    df_clean = df.drop(columns=monthly_cols)
    
    print(f"\n✅ Removed {len(monthly_cols)} monthly columns")
    print(f"✅ Kept seasonal columns only")
    
    return df_clean


df_seasonal_only = replace_monthly_with_seasonal(df.copy())
df_seasonal_only.to_csv(r"D:\S3\rapport data_mining\newdata\climate_fire_soil_seasonal.csv", index=False)

# ============================================================================
# METHOD 3: CUSTOM SEASON DEFINITIONS (NORTH AFRICA)
# ============================================================================

def create_custom_seasonal_features(df):
    """
    Create seasonal features with custom definitions for North Africa climate.
    
    North Africa seasons:
    - Hot Dry Season: Jun, Jul, Aug (summer)
    - Cool Wet Season: Dec, Jan, Feb (winter)
    - Transition Spring: Mar, Apr, May
    - Transition Autumn: Sep, Oct, Nov
    """
    
    custom_seasons = {
        'hot_dry': ['jun', 'jul', 'aug'],
        'cool_wet': ['dec', 'jan', 'feb'],
        'spring_transition': ['mar', 'apr', 'may'],
        'autumn_transition': ['sep', 'oct', 'nov']
    }
    
    for var_type in ['tmin', 'tmax', 'prec']:
        for season_name, months in custom_seasons.items():
            season_cols = [f"{var_type}_{month}" for month in months 
                          if f"{var_type}_{month}" in df.columns]
            
            if season_cols:
                df[f"{var_type}_{season_name}"] = df[season_cols].mean(axis=1)
                print(f"✅ Created {var_type}_{season_name}")
    
    return df

df_custom = create_custom_seasonal_features(df.copy())
df_custom.to_csv(r"D:\S3\rapport data_mining\newdata\climate_fire_soil_seasonal.csv", index=False)

# ============================================================================
# COMPARISON & ANALYSIS
# ============================================================================

print("\n" + "="*70)
print("COMPARISON: MONTHLY vs SEASONAL")
print("="*70)

# Count features
monthly_count = len([c for c in df.columns if any(m in c for m in 
                    ['jan', 'feb', 'mar', 'apr', 'may', 'jun',
                     'jul', 'aug', 'sep', 'oct', 'nov', 'dec'])])

seasonal_count = len([c for c in df_seasonal.columns if any(s in c for s in 
                     ['winter', 'spring', 'summer', 'autumn'])])

seasonal_only_count = len([c for c in df_seasonal_only.columns if any(s in c for s in 
                          ['winter', 'spring', 'summer', 'autumn'])])

print(f"\nOriginal dataset:")
print(f"  Monthly features: {monthly_count}")
print(f"  Total columns: {len(df.columns)}")

print(f"\nWith seasonal added (METHOD 1):")
print(f"  Monthly features: {monthly_count}")
print(f"  Seasonal features: {seasonal_count}")
print(f"  Total columns: {len(df_seasonal.columns)}")

print(f"\nSeasonal only (METHOD 2):")
print(f"  Seasonal features: {seasonal_only_count}")
print(f"  Total columns: {len(df_seasonal_only.columns)}")
print(f"  Reduction: {len(df.columns) - len(df_seasonal_only.columns)} columns")

# Show new column names
print(f"\n📋 New Seasonal Columns Created:")
seasonal_cols = [c for c in df_seasonal.columns if any(s in c for s in 
                ['winter', 'spring', 'summer', 'autumn'])]
for col in sorted(seasonal_cols):
    print(f"  - {col}")


## **📊 Output Files:**
"""
1. **`climate_fire_soil_seasonal.csv`** - Original + seasonal (both monthly & seasonal)
2. **`climate_fire_soil_seasonal_only.csv`** - Only seasonal (monthly removed)
3. **`climate_fire_soil_custom_seasons.csv`** - Custom North Africa seasons

## **🌡️ What Gets Created:**

### **Standard Seasons:**
```
tmin_winter, tmin_spring, tmin_summer, tmin_autumn
tmax_winter, tmax_spring, tmax_summer, tmax_autumn
prec_winter, prec_spring, prec_summer, prec_autumn"""
````

````python
import rasterio
import pandas as pd
from rasterio.transform import rowcol
import numpy as np
import glob

# --- 1. DEFINE FILE PATHS ---
climate_fire_file = r"D:\S3\data_mining\projet\final_climate_fire_soil_data.csv"
soil_file = r"D:\S3\data_mining\projet\donnees_elevation_coordonnees.csv" 
tif_folder = r"D:\S3\data_mining\projet\clipped_imputed_data2\\" 
tif_files = sorted(glob.glob(tif_folder + "*_imputed_local.tif"))

# --- 2. LOAD DATA ---
climate_fire_df = pd.read_csv(climate_fire_file)
soil_df = pd.read_csv(soil_file)

# Get the Transformation matrix from the first climate TIF file
with rasterio.open(tif_files[0]) as src:
    transform = src.transform

# --- 3. ADD ROW AND COL TO CLIMATE_FIRE_DF ---
# Calculate 'row' and 'col' for the climate_fire data using the TIF transform
print("Adding row and col to climate_fire data...")
rows, cols = rowcol(transform, climate_fire_df['longitude'], climate_fire_df['latitude'])
climate_fire_df['row'] = rows
climate_fire_df['col'] = cols

# --- 4. MAP SOIL COORDINATES TO GRID INDICES ---
# Calculate 'row' and 'col' for the soil data using the TIF transform
print("Adding row and col to soil data...")
rows, cols = rowcol(transform, soil_df['longitude'], soil_df['latitude'])
soil_df['row'] = rows
soil_df['col'] = cols

# --- 5. PREPARE SOIL DATA FOR MERGE ---
# Keep all soil property columns and drop the redundant float coordinates
soil_cols_to_keep = [
    col for col in soil_df.columns 
    if col not in ['longitude', 'latitude']
]
# Ensure we only have unique grid cells for merging
soil_df_final = soil_df[soil_cols_to_keep].drop_duplicates(subset=['row', 'col'])

print(f"Climate fire data shape: {climate_fire_df.shape}")
print(f"Soil data shape: {soil_df_final.shape}")
print(f"Climate fire unique grid cells: {climate_fire_df[['row', 'col']].drop_duplicates().shape[0]}")
print(f"Soil unique grid cells: {soil_df_final[['row', 'col']].drop_duplicates().shape[0]}")

# --- 6. PERFORM THE FINAL MERGE ---
# Use a LEFT JOIN to keep all rows in the climate_fire_df
final_combined_df = climate_fire_df.merge(
    soil_df_final,
    on=['row', 'col'],
    how='left'
)

# --- 7. CLEANUP AND SAVE ---
print(f"Final Combined DataFrame size: {len(final_combined_df)}")
print(f"Number of pixels with missing soil data: {final_combined_df['HWSD2_SMU_ID'].isna().sum()}")

# Check which columns have missing data
missing_data = final_combined_df.isnull().sum()
print("\nMissing data per column:")
print(missing_data[missing_data > 0])

# Save the final, complete dataset
final_combined_df.to_csv(r"D:\S3\data_mining\projet\final_climate_fire_soil_elevation_data.csv", index=False)
print("Final dataset saved successfully!")
````
